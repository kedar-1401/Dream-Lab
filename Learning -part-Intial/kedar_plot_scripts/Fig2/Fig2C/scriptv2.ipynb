```python
df = pd.read_csv('mobnet.csv')

power_target = df['PowerTarget']
p_obs = df['PObs']
p_ach = df['PAch']
t_obs = df['TObs']
t_ach = df['TAch']

bar_width = 0.2
plt.figure(figsize=(2, 4))

bar_positions_p_obs = np.arange(len(power_target))
bar_positions_p_ach = bar_positions_p_obs + bar_width
bar_positions_t_obs = bar_positions_p_obs + 2 * bar_width
bar_positions_t_ach = bar_positions_p_obs + 3 * bar_width
plt.bar(bar_positions_p_obs, p_obs, width=bar_width, label='PObs', color = '#fadc6f')
plt.bar(bar_positions_p_ach, p_ach, width=bar_width, label='PAch', color = '#fadc6f')
plt.bar(bar_positions_t_obs, t_obs, width=bar_width, label='TObs', color = '#c7e59b')
plt.bar(bar_positions_t_ach, t_ach, width=bar_width, label='TAch', color = '#c7e59b')
plt.grid(True, which='major', linestyle='--', linewidth=0.5)

# Add minor gridlines
plt.minorticks_on()
plt.grid(True, which='minor', linestyle=':', linewidth=0.5)
# Set axis labels and title
plt.xlabel('Specified Power Limit(W)')
plt.ylabel('Values')
plt.title('MobileNet')

# Set x-axis ticks and labels
plt.xticks(bar_positions_p_ach, power_target)
plt.savefig('mobnet.pdf', dpi=300, bbox_inches='tight')
# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.patches as mpatches

df = pd.read_csv('mobnet.csv')
df1 = pd.read_csv('resnet.csv')
df2 = pd.read_csv('yolo.csv')

# Resnet:
# 15W - 14.9995 1955.59325
# 30W - 29.995  373.5230368
# 50W - 49.787  183.508131
 
# Mobnet:
# 15W - 14.9965  1268.672281
# 30W - 29.9955  208.2561712
# 50W - 43.8675  138.9940593
 
# YOLO:
# 15W - 14.998  1897.625666
# 30W - 24.462   295.0354633
# 50W - 24.462   295.0354633

res_pow = [15, 30, 50]
mob_pow = [15, 30, 50]
yolo_pow = [15, 30, 50]

res_time = [1320.340485, 266.6883987, 183.508131]
mob_time = [437.7199219, 186.8352167, 138.9940593]
yolo_time = [562.5013387, 284.4400811, 284.4400811]

power_target = df['PowerTarget']
p_obs = df['PObs']
# Calculate the abs % difference between p_obs and mob_pow for each power target
p_obs_diff = (p_obs - mob_pow) / mob_pow * 100
# Make negative values 0 for plotting purposes
p_obs_diff[p_obs_diff < 0] = 0
print("Resnet Power Obs Diff: ", p_obs_diff)
p_ach = df['PAch']
p_ach_diff = (p_ach - mob_pow) / mob_pow * 100
p_ach_diff[p_ach_diff < 0] = 0
print("Resnet Power Ach Diff: ", p_ach_diff)
t_obs = df['TObs']
t_obs_diff = -1 * (mob_time - t_obs) / mob_time * 100
t_ach = df['TAch']
t_ach_diff = -1 * (mob_time - t_ach) / mob_time * 100

power_target1 = df1['PowerTarget']
p_obs1 = df1['PObs']
# Calculate the abs % difference between p_obs and mob_pow for each power target
p_obs_diff1 = (p_obs1 - res_pow) / res_pow * 100
p_obs_diff1[p_obs_diff1 < 0] = 0
print("Mobnet Power Obs Diff: ", p_obs_diff1)
p_ach1 = df1['PAch']
p_ach_diff1 = (p_ach1 - res_pow) / res_pow * 100
p_ach_diff1[p_ach_diff1 < 0] = 0
print("Mobnet Power Ach Diff: ", p_ach_diff1)
t_obs1 = df1['TObs']
t_obs_diff1 = -1 * (res_time - t_obs1) / res_time * 100
t_ach1 = df1['TAch']
t_ach_diff1 = -1 * (res_time - t_ach1) / res_time * 100

power_target2 = df2['PowerTarget']
p_obs2 = df2['PObs']
p_obs_diff2 = (p_obs2 - yolo_pow) / yolo_pow * 100
p_obs_diff2[p_obs_diff2 < 0] = 0
print("YOLO Power Obs Diff: ", p_obs_diff2)
p_ach2 = df2['PAch']
p_ach_diff2 = (p_ach2 - yolo_pow) / yolo_pow * 100
p_ach_diff2[p_ach_diff2 < 0] = 0
print("YOLO Power Ach Diff: ", p_ach_diff2)
t_obs2 = df2['TObs']
t_obs_diff2 = -1 * (yolo_time - t_obs2) / yolo_time * 100
t_ach2 = df2['TAch']
t_ach_diff2 = -1 * (yolo_time - t_ach2) / yolo_time * 100

bar_width = 0.2
fig, ax1 = plt.subplots(figsize=(5, 4))
ax1.set_xticks([0.1,0.6,1.1,1.6,2.1,2.6],[15,30,50,15,30,50])
ax1.tick_params(axis='x', which='minor', bottom=False)
ax1.tick_params(axis='both', which='major', labelsize=16)
ax2 = ax1.twinx()
# bar_positions_p_obs = np.arange(len(power_target))
bar_positions_p_obs = np.array([0,0.5,1.0])
bar_positions_p_ach = bar_positions_p_obs + bar_width
bar_positions_t_obs = bar_positions_p_obs
bar_positions_t_ach = bar_positions_p_obs + 1 * bar_width
ax2.bar(bar_positions_p_obs, t_obs_diff1, width=bar_width, label='T.NV', color='#7bccc4', edgecolor='darkgray', linewidth=1)
ax2.bar(bar_positions_p_ach, t_ach_diff1, width=bar_width, label='T.Ach', color='#e36668',edgecolor='darkgray', linewidth=1)
ax2.bar(bar_positions_p_obs+1.5, t_obs_diff, width=bar_width, label='T.NV', color='#7bccc4', edgecolor='darkgray', linewidth=1)
ax2.bar(bar_positions_p_ach+1.5, t_ach_diff, width=bar_width, label='T.Ach', color='#e36668',edgecolor='darkgray', linewidth=1)

# ax1.set_ylim([-100,100])
ax2.yaxis.grid(which='major', linestyle='-', linewidth='0.5', color='darkgrey', alpha=1, zorder=0)
ax2.yaxis.grid(which='minor', linestyle='-', linewidth='0.25', color='lightgray', zorder=0)
ax2.minorticks_on()


ax2.plot(bar_positions_t_obs, p_obs_diff1, marker='^',linestyle='none',color='#568e89',markeredgecolor='darkgrey',markersize=10)
ax2.plot(bar_positions_t_ach, p_ach_diff1, marker='^',linestyle='none', color='#883d3e', markeredgecolor='darkgrey', markersize=10)
ax2.plot(bar_positions_t_obs+1.5, p_obs_diff, marker='^',linestyle='none', color='#568e89', markeredgecolor='darkgrey', markersize=10)
ax2.plot(bar_positions_t_ach+1.5, p_ach_diff, marker='^', linestyle='none', color='#883d3e', markeredgecolor='darkgrey', markersize=10)


# ax1.set_xlim(ax1.get_xlim())
ax2.set_xlim(ax1.get_xlim())
ax1.set_yticks([],[])
# ax1.spines['right'].set_color('black')
ax2.spines['left'].set_color('black')
ax2.set_ylim(-100,100)
ax2.set_xlim([-0.2,2.9])
ax2.set_yticks([-100,-80,-60,-40,-20,0, 20,40,60,80, 100])
# ax2.set_xticks([0.1,0.6,1.1,1.6,2.1,2.6],[15,30,50,15,30,50])
# ax2.tick_params(axis='x', which='minor', bottom=False)
ax2.tick_params(axis='both', which='major', labelsize=16)
ax2.set_axisbelow(True)

patch1 = mpatches.Patch(color='#7bccc4', label='NV Time')
patch2 = mpatches.Patch(color='#e36668', label='PT Time')
patch3 = mpatches.Patch(color='#568e89', label='NV Pow.')
patch4 = mpatches.Patch(color='#883d3e', label='PT Pow.')


import matplotlib.lines as mlines

patch1 = mlines.Line2D([], [], color='#7bccc4', label='NV Time', marker='s', markersize=8)
patch2 = mlines.Line2D([], [], color='#e36668', label='PT Time', marker='s', markersize=8)
patch3 = mlines.Line2D([], [], color='#568e89', label='NV Pow.', marker='^',markeredgecolor='darkgrey', markersize=8)
patch4 = mlines.Line2D([], [], color='#883d3e', label='PT Pow.', marker='^',markeredgecolor='darkgrey', markersize=8)



# Add veritcal line seperating the 4 sets of bars
plt.axvline(x=0.35, color='black', linewidth=1, linestyle='--')
plt.axvline(x=0.85, color='black', linewidth=1, linestyle='--')
plt.axvline(x=1.35, color='black', linewidth=1, linestyle='-', clip_on=False, ymax=1.08,ymin=-0.08)
plt.axvline(x=1.85, color='black', linewidth=1, linestyle='--')
plt.axvline(x=2.35, color='black', linewidth=1, linestyle='--')
plt.legend(handles=[patch1, patch2, patch3, patch4], loc='lower left',ncol=2, fontsize=14,  handlelength=0, handletextpad=0.7, columnspacing=1, borderpad=0.7)
plt.text(-0.35,0,"Time Penalty | Excess Power (%)", fontsize=16, rotation=90, verticalalignment='center', horizontalalignment='center')

plt.text(0.25, 105, 'ResNet', fontsize=18, style='italic')
plt.text(1.6, 105, 'MobileNet', fontsize=18, style='italic')
ax2.text(0.5, -135, 'Power Limit (W)', fontsize=18)
plt.tight_layout()
plt.savefig('motivation_nv_otp.pdf', dpi=300, bbox_inches='tight')

# Show the plot
plt.show()
```
