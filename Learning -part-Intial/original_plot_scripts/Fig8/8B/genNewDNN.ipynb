```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Sample data
plotdata_time = pd.DataFrame({
    "NN": [12.3, 15.1],
    "PT": [12.5, 15.6]
}, index=["LSTM", "BERT"])
plotdata_power = pd.DataFrame({
    "NN": [9.1, 9.3],
    "PT": [6.3, 5.8]
}, index=["LSTM", "BERT"])

# Plotting
fig, ax = plt.subplots(figsize=(4, 2.5))

# Bar plot on the left y-axis
bar_width = 0.15
bar_positions_time = np.arange(len(plotdata_time.index))
# ax.bar(bar_positions_time - bar_width / 4, plotdata_time["NN - Time"], width=bar_width / 2, label="NN - Time", color='#fadc6f', edgecolor='grey')
# ax.bar(bar_positions_time + bar_width / 4, plotdata_time["TL - Time"], width=bar_width / 2, label="TL - Time", color='#e36668', edgecolor='grey')
ax.bar((0,0.45), plotdata_time["NN"], width=bar_width, label="NN", color='#fadc6f', edgecolor='darkgrey')
ax.bar((0.15,0.6), plotdata_time["PT"], width=bar_width, label="PT", color='#e36668', edgecolor='darkgrey')

# Create a secondary y-axis on the right
ax2 = ax.twinx()

ax2.plot((0,0.45) , plotdata_power["NN"], color='#a09040', marker = '^', linestyle = 'none', markeredgecolor = 'darkgrey', markersize = 10)
ax2.plot((0.15,0.6) , plotdata_power["PT"], color='#762e30', marker = '^', linestyle = 'none', markeredgecolor = 'darkgrey', markersize = 10)

# Bar plot on the right y-axis
# bar_positions_power = np.arange(len(plotdata_power.index)) + len(bar_positions_time)
# ax2.bar(bar_positions_power - bar_width / 4, plotdata_power["NN - Power"], width=bar_width / 2, label="NN - Power", color='#fadc6f', edgecolor='grey')
# ax2.bar(bar_positions_power + bar_width / 4, plotdata_power["TL - Power"], width=bar_width / 2, label="TL - Power", color='#e36668', edgecolor='grey')
 
# Add a vertical line to separate the two dataframes
# ax.axvline(x=len(bar_positions_time) - 0.5, color='black', linestyle='-', linewidth=1)
 
# Set labels and title with increased font size
ax.set_ylabel('Time MAPE (%)', fontsize=14)
ax2.set_ylabel('Power MAPE (%)', fontsize=14)
#ax.set_xlabel('Models', fontsize=16)
#ax.set_title('Grouped Bar Plot for LSTM and BERT', fontsize=16)
 
# Set legends with increased font size
#ax2.legend(loc='upper left')
#ax.legend(loc='upper right')
plt.legend(["NN", "PT"], loc = "upper center", fontsize = 16, ncols=2)# Set x-axis ticks and labels with increased font size
ax.set_xticks([0.1,0.55],["LSTM", "BERT"], fontsize=16)
# ax.set_xticks(np.concatenate([bar_positions_time, bar_positions_power]))
# ax.set_xticklabels(plotdata_time.index.tolist() + plotdata_power.index.tolist(), fontsize=11.5)
ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=True)

# Set y-axis ticks and labels with increased font size
ax.tick_params(axis='y', labelsize=14)
ax2.tick_params(axis='y', labelsize=14)
ax.yaxis.grid(which='major', linestyle='-', linewidth='0.5', color='darkgrey', alpha=1)
ax.set_axisbelow(True)
ax.yaxis.grid(which='minor', linestyle='-', linewidth='0.25', color='lightgray')
# ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=True)
 
ax.minorticks_on()

# Set y-axis limits
ax.set_ylim(0, 20)
ax2.set_ylim(0, 20)
plt.tight_layout()
# Show the plot
plt.savefig("gen_new_DNN_plot.pdf", dpi = 600, format = "pdf")
plt.savefig("gen_new_DNN_plot.png", dpi = 600, format = "png")
```
