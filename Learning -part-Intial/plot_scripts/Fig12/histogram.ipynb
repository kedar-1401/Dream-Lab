```python
def convert_data(input_string):
    # Split the input string into components
    parts = input_string.split('\t')

    # Extract data for each bar color
    blue_data = [float(value) for value in parts[2].split(',')]
    orange_data = [float(value) for value in parts[3].split(',')]
    green_data = [float(value) for value in parts[4].split(',')]

    # Create the output format
    output = f"blue_bar = ({blue_data[0]}, {orange_data[0]}, {green_data[0]})\n"
    output += f"orange_bar = ({blue_data[1]}, {orange_data[1]}, {green_data[1]})\n"
    output += f"green_bar = ({blue_data[2]}, {orange_data[2]}, {green_data[2]})"

    return output
```
