```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample data
plotdata_time = pd.DataFrame({
    "NN": [21.1, 20.5],
    "Cross Device TL": [12.1, 14.3]
}, index=["RNet", "MNet"])

plotdata_power = pd.DataFrame({
    "NN": [18.3, 15.2],
    "Cross Device TL": [10.8, 8.6],
}, index=["RNet", "MNet"])

# Plotting
fig, ax = plt.subplots(figsize=(4, 2.5))

# Bar plot on the left y-axis (Time)
bar_width = 0.2
bar_positions_time = np.arange(len(plotdata_time.index))
# ax.bar(bar_positions_time - bar_width*5, plotdata_time["NN"], width=bar_width, label="NN", color='#fadc6f', edgecolor='darkgrey')
# ax.bar(bar_positions_time, plotdata_time["Intra Device TL"], width=bar_width, label="Intra Device TL - Time", color='#e36668', edgecolor='darkgrey')
# ax.bar(bar_positions_time , plotdata_time["Cross Device TL"], width=bar_width, label="Cross Device TL", color='#e36668', edgecolor='darkgrey')
ax.bar((0,0.6), plotdata_time["NN"], width=bar_width, label="NN", color='#fadc6f', edgecolor='darkgrey')
ax.bar((0.2,0.8), plotdata_time["Cross Device TL"], width=bar_width, label="Cross Device PT", color='#e36668', edgecolor='darkgrey')


# Create a secondary y-axis on the right
ax2 = ax.twinx()

# Bar plot on the right y-axis (Power)
bar_positions_power = np.arange(len(plotdata_power.index)) + len(bar_positions_time)
# ax2.plot(bar_positions_power - bar_width*15, plotdata_power["NN"], color='#fadc6f', marker = '^', linestyle = 'none', markeredgecolor = 'darkgrey', markersize = 10)
# ax2.plot(0, plotdata_power["NN"], color='#fadc6f', marker = '^', linestyle = 'none', markeredgecolor = 'darkgrey', markersize = 10)
ax2.plot((0,0.6) , plotdata_power["NN"], color='#a09040', marker = '^', linestyle = 'none', markeredgecolor = 'darkgrey', markersize = 10)
ax2.plot((0.2,0.8) , plotdata_power["Cross Device TL"], color='#762e30', marker = '^', linestyle = 'none', markeredgecolor = 'darkgrey', markersize = 10)



# Set labels and title with increased font size
ax.set_ylabel('Time MAPE (%)', fontsize=14)
ax2.set_ylabel('Power MAPE (%)', fontsize=14)

# Set legends with increased font size

plt.legend(["NN", "PT"], loc = "lower center", fontsize = 16, ncols=2)

# Set x-axis ticks and labels with increased font size
# ax.set_xticks(np.concatenate([bar_positions_time, bar_positions_power]))
ax.set_xticks([0.1,0.7],["RNet", "MNet"], fontsize=16)
# ax.set_xticklabels(plotdata_time.index.tolist() + plotdata_power.index.tolist(), fontsize=14)
ax.tick_params(axis='x', which='both', bottom=False, top=False, labelbottom=True)

# Set y-axis ticks and labels with increased font size
ax.tick_params(axis='y', labelsize=14)
ax2.tick_params(axis='y', labelsize=14)

ax.yaxis.grid(which='major', linestyle='-', linewidth='0.5', color='darkgrey', alpha=1)
ax.set_axisbelow(True)
ax.yaxis.grid(which='minor', linestyle='-', linewidth='0.25', color='lightgray')
ax.minorticks_on()

# Set y-axis limits
ax.set_ylim(0, 25)
ax2.set_ylim(0, 25)

plt.tight_layout()
# Show the plot
plt.savefig("gen_new_Device_plot.pdf", dpi=300, format="pdf")
plt.savefig("gen_new_Device_plot.png", dpi=300, format="png")
plt.show()
```
