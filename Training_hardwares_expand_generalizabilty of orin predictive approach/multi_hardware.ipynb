```python
models = ['ResNet', 'MobileNet', 'YOLO', 'LSTM', 'BERT']
colors_3090 = '#be4986'
colors_orin3 = '#fadc6f'
colors_A5000 = '#e36668'
colors_rpi5 = '#88c999'
# Width of a single bar
bar_width = 0.2

# Positions of the bars on the x-axis
r1 = range(len(models))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]
r4 = [x + bar_width for x in r3]

# Plotting the grouped bar plot
plt.figure(figsize=(18, 8))
bars1 = plt.bar(r1, average_epoch_times_3090, color=colors_3090, width=bar_width, edgecolor='grey', label='3090')
bars2 = plt.bar(r2, average_epoch_times_A5000, color=colors_A5000, width=bar_width, edgecolor='grey', label='A5000')
bars3 = plt.bar(r3, average_epoch_times_orin3, color=colors_orin3, width=bar_width, edgecolor='grey', label='Orin*')
bars4 = plt.bar(r4, average_epoch_times_RPI5, color=colors_rpi5, width=bar_width, edgecolor='grey', label='RPi5')

# Setting the y-axis to logarithmic scale
plt.yscale('log')
plt.ylim(10, 26000)

# Adding labels to the plot
plt.ylabel('Average epoch time in s (Log)', fontsize=28)

# Setting x-ticks and adding model names as labels
plt.xticks([r + 1.5 * bar_width for r in r1], models, fontsize=26)

# Adding grid lines
plt.grid(True, axis='y', linestyle='-', alpha=0.7)
plt.minorticks_on()
plt.grid(which='minor', axis='y', linestyle=':', alpha=0.4)
plt.gca().set_axisbelow(True)  

# Adding a legend
plt.legend(fontsize=20)

plt.yticks(fontsize=20)

# Adding value labels on top of the bars
for bars in [bars1, bars2, bars3, bars4]:
    for bar in bars:
        yval = bar.get_height()
        if yval > 20000:
            plt.text(bar.get_x() + bar.get_width() / 2, 21000, round(yval, 2), ha='center', color='red', fontsize=17, fontweight='bold')
        elif yval == 0.0:
            plt.text(bar.get_x() + bar.get_width() / 2, yval + 10.5, 'DNR', ha='center', color='red', fontsize=17, fontweight='bold')
        else:
            plt.text((bar.get_x() + bar.get_width() / 2), yval + 5, round(yval), ha='center', fontsize=17, fontweight='bold')

# Showing the plot
plt.savefig('Multi_Hardware_Training_Plot.pdf', format='pdf', dpi=300, bbox_inches='tight')
plt.show()
```
