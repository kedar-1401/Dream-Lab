```python
from turtle import pos
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10, 6))


ax.xaxis.grid(False)
ax.minorticks_on()
bar_width = 0.27  # Decreased the bar width for better visualization

bar_positions1 = new_df['power_budget'] - 1.5 * bar_width
bar_positions2 = new_df['power_budget'] - 0.5 * bar_width
bar_positions3 = new_df['power_budget'] + 0.5 * bar_width
bar_positions4 = new_df['power_budget'] + 1.5 * bar_width
bar_positions5 = new_df['power_budget'] + 2.5 * bar_width


bars1 = ax.bar(bar_positions1, new_df['random_time'], bar_width, color='#e36668', alpha=1, label='RND', edgecolor='#606060')
bars2 = ax.bar(bar_positions2, new_df['nn_time'], bar_width, color='#fadc6f', alpha=1, label='NN250', edgecolor='#606060')
bars3 = ax.bar(bar_positions3, new_df['observed_time_bs'], bar_width, color='#c7e59b', alpha=1, label='BS', edgecolor='#606060')
bars4 = ax.bar(bar_positions4, new_df['algo_time'], bar_width, color='#84b070', alpha=1, label='GMD', edgecolor='#606060')
bars5 = ax.bar(bar_positions5, new_df['optim_time'], bar_width, color='#61b1ff', alpha=1, label='OPT', edgecolor='#606060')


ax2 = ax.twinx()
ax2.plot(bar_positions3, df4['observed_power_nn'], marker='v', markersize=8, linestyle='', color='#eb9395', alpha=1, label='NN250 Power', markeredgecolor='#713334')# ax2.plot(bar_positions4, df4['observed_power_random'], marker='x', markersize=4, linestyle='', color='#ff7f0e', alpha=1, label='Power Random')
ax2.plot(bar_positions3, new_df['power_budget'], marker='D', markersize=8, linestyle='', color='red', alpha=1, label='Power Budget')

for bar, value in zip(bars3, new_df['optim_time']):
    if value == 0:
        ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, 'no solution for optimal', ha='center', va='bottom', rotation=90, fontsize=5)

# for bar, value in zip(bars2, new_df['algo_time']):
#     if value == 0:
#         ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1, 'no solution for gradient multi dim', ha='center', va='bottom', rotation=90, fontsize=5)


ax.set_xlabel('Power Budget (W)', fontsize=22)
ax.set_ylabel('Train Minibatch Time (ms)', fontsize=22)
ax2.set_ylabel('Power (W)', fontsize=22)

ax.legend(ncols=5, fontsize=14, loc='upper center')
ax.set_axisbelow(True)

# Set xticks to power_budget values
ax.set_xticks(power_fil_list)
ax.set_yticks(np.arange(0, 160, 20))
ax.tick_params(axis='x', which='minor', bottom=False)
ax.set_xticklabels(power_fil_list, fontsize=18)
ax.set_yticklabels([0, 20, 40, 60, 80, 100, 120, 140], fontsize=18)


ax.set_xlim(left=27)
ax.set_xlim(right=43)

ax.set_ylim(0,140)
ax2.set_ylim(0, 70)
ax2.set_yticklabels([0, 10, 20, 30, 40, 50, 60, 70], fontsize=18)

ax.minorticks_on()

# Add major gridlines
ax.yaxis.grid(which='major', linestyle='-', linewidth='0.5', color='#707070', alpha=1)

# Add horizontal minor gridlines
ax.yaxis.grid(which='minor', linestyle='-', linewidth='0.25', color='lightgray')

# Remove minor ticks on the x-axis
ax.set_xticks([], minor=True)
ax2.legend(loc='lower center', fontsize=14, ncols=2)
plt.savefig('trainZoomIn.pdf', format='pdf',dpi=300, bbox_inches='tight')

plt.show()
```
