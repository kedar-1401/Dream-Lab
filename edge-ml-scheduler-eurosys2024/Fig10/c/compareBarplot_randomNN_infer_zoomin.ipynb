```python
# merge all df on arr_rate, power_budget, time_budget
df = pd.merge(gmd, nnRan250, on=['arr_rate', 'power_budget', 'time_budget'])
df = pd.merge(df, nnRan50, on=['arr_rate', 'power_budget', 'time_budget'])
```

```python
df.columns
```

```python
df['time_cond1_algo'] = df['infer_time_algo'] + (df['bs_algo']/df['arr_rate'])*1000
df['time_cond1_optim'] = df['infer_time_optim'] + (df['bs_optim']/df['arr_rate'])*1000
df['time_cond1_random'] = df['infer_time_random_y'] + (df['bs_random_y']/df['arr_rate'])*1000
df['time_cond1_nn'] = df['infer_time_nn_x'] + (df['bs_nn_x']/df['arr_rate'])*1000
```
